```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

from report_results import *
# 设置中文字体
font = FontProperties(fname="/Users/k/Library/Fonts/LXGWWenKai-Regular.ttf")


```

```python


def plot_multiple_accuracies(exp_dict):
    plt.figure(figsize=(10, 6))  # 设置图像尺寸
    for label, exp_name in exp_dict.items():
        full_path = f'./results/{exp_name}/accuracies_losses_valid.csv'
        df = pd.read_csv(full_path)
        # val_loss = df['val_loss'][:57]
        # iteration = df['iteration'][:57]
        val_loss = df['val_loss']
        iteration = df['iteration']
        plt.plot(iteration, val_loss, label=label, linewidth=2)

    plt.xlabel('迭代步数', fontsize=14, fontproperties=font)  # 设置x轴标签和字体大小
    plt.ylabel('损失', fontsize=14, fontproperties=font)  # 设置y轴标签和字体大小
    # plt.title('模型准确率比较', fontsize=16, fontproperties=font)  # 设置标题和字体大小
    plt.legend(fontsize=12, prop=font)  # 设置图例和字体大小
    plt.grid(True)  # 添加网格线
    plt.xticks(fontsize=12)  # 设置x轴刻度字体大小
    plt.yticks(fontsize=12)  # 设置y轴刻度字体大小
    plt.tight_layout()  # 自动调整子图参数，使之填充整个图像区域
    # plt.savefig(f'./results/{exp_name}/accuracy_comparison.png', dpi=300)  # 保存图像，提高分辨率
    plt.show()
    plt.close()



# 调用函数绘制图像
# plot_multiple_accuracies(SPTS_exp_dict)
plot_multiple_accuracies(TFIM_exp_dict)

```
